# Delayed reporting example

In this script want to create a quick visual for describing the effects of adding of a 'reporting delay threshold' for the heat wave analysis. 

## Set up

```python
#--- Libraries
import pandas  as pd                # stats packages
import numpy   as np                # linear algebra packages

import matplotlib.pyplot as plt     # ploting packages
import seaborn as sns               # more plotting routines

#--- Configure plotting environments/defaults
# use 'cartoon-style'
plt.xkcd()
# use white background 
sns.set_style('white')
# set color choices
c = sns.color_palette('deep')

# show plots in notebook 
% matplotlib inline
```

## Create data

Invent some time series of meteorologically defined warm events, and reported humanitarian disasters. 

```python
#--- Invent data
# choose dates
dates = pd.date_range(start='2050-05-01',end='2050-08-31')
# set series of meteorologicaly defined heat waves
met_events = pd.Series(index = dates)
# choose some dates to be 'defined heat waves'
met_events['2050-05-12':'2050-05-19'] = 1
met_events['2050-07-20':'2050-08-2'] = 1
dummy_line = pd.Series(np.repeat(1,len(dates)),index=dates)
# set series of reported heat induced crisis
rep_events = pd.Series(index=dates)
# choose some dates to be 'reported disasters' 
rep_events['2050-05-24'] = -1
rep_events['2050-07-22'] = -1
rep_events['2050-08-014'] = -1
# list selected dates
rep_dates = rep_events.index[rep_events==-1]
```

## Show effect of delay threshold

Show how the warm events and the reported disasters line up, then how they line up if allowing that the reports may have come in after the fact.

```python
#--- Match reports to warm events
# create figure
fig, chart = plt.subplots(3,figsize=(12,18))
# show warm events
_ = chart[0].plot(met_events.index,met_events,color=c[2],lw=10)
# draw a line showing intersection with reports
map(lambda x: chart[0].axvline(x=rep_dates[x],ymin=-1,ymax=0.9,linestyle='--',color='gray',alpha=0.5), 
    range(3))
# show date of reports
map(lambda x: chart[0].text(rep_dates[x],-1,r'$\blacktriangle$',fontsize=50,color='orange',horizontalalignment='center'),
    range(3))
map(lambda x: chart[0].text(rep_dates[x],-0.95,r'$!$',fontsize=20,color='black',horizontalalignment='center'),
    range(3))
# show full date range
_ = chart[0].set_xlim('2050-05-01','2050-08-31')
# adjust y-axis span
_ = chart[0].set_ylim(-2,2)
# set the dates between the two records
_ = chart[0].spines['bottom'].set_position('center')
# only label values in three places
_ = chart[0].set_yticks([-1,0,1])
# name indicated data
_ = chart[0].set_yticklabels(['reports ','dates','warm events'],fontsize=14)
# label number of disasters detected
_ = chart[0].text('2050-06-07',1.2,'Delay threshold = 0', style='italic',
               bbox={'facecolor':c[4], 'alpha':0.4, 'pad':20})
_ = chart[0].text('2050-06-07',0.75,'Disasters detected = 1', style='italic',
               bbox={'facecolor':c[4], 'alpha':0.4, 'pad':20})
# remove left axis line
sns.despine(bottom=False,left=True,top=True,ax=chart[0])

#--- Match reports to warm events, with 5 day threshold
# show warm events
_ = chart[1].plot(met_events.index,met_events,color=c[2],lw=10)
# draw a line showing intersection with reports
map(lambda x: chart[1].axvline(x=rep_dates[x],ymin=-1,ymax=0.9,linestyle='--',color='gray',alpha=0.5), 
    range(3))
# draw a line showing intersection with reports
map(lambda x: chart[1].axvline(x=rep_dates[x]-pd.Timedelta(5,'D'),ymin=-1,ymax=0.9,linestyle='--',color='gray',alpha=0.5), 
    range(3))
map(lambda x: chart[1].fill_between(x=pd.date_range(rep_dates[x]-pd.Timedelta(5,'D'),rep_dates[x]),
                                 y1=-2,y2=1.5,color='gray',alpha=0.2),
    range(3))
# show date of reports
map(lambda x: chart[1].text(rep_dates[x],-1,r'$\blacktriangle$',fontsize=50,color='orange',horizontalalignment='center'),
    range(3))
map(lambda x: chart[1].text(rep_dates[x],-0.95,r'$!$',fontsize=20,color='black',horizontalalignment='center'),
    range(3))
# show full date range
_ = chart[1].set_xlim('2050-05-01','2050-08-31')
# adjust y-axis span
_ = chart[1].set_ylim(-2,2)
# set the dates between the two records
_ = chart[1].spines['bottom'].set_position('center')
# only label values in three places
_ = chart[1].set_yticks([-1,0,1])
# name indicated data
_ = chart[1].set_yticklabels(['reports','dates','warm events'],fontsize=14)
# label number of disasters detected
_ = chart[1].text('2050-06-07',1.2,'Delay threshold = 5 days', style='italic',
                  bbox={'facecolor':c[4], 'alpha':0.4, 'pad':20})
_ = chart[1].text('2050-06-07',0.75,'Disasters detected = 2', style='italic',
                  bbox={'facecolor':c[4], 'alpha':0.4, 'pad':20})
# remove left axis line, add top line
sns.despine(bottom=False,left=True,top=False,ax=chart[1])
# set color of dividing line
chart[1].spines['top'].set_color(c[0])

#--- Match reports to warm events, with 12 day threshold
# show warm events
_ = chart[2].plot(met_events.index,met_events,color=c[2],lw=10)
# draw a line showing intersection with reports
map(lambda x: chart[2].axvline(x=rep_dates[x],ymin=-1,ymax=0.9,linestyle='--',color='gray',alpha=0.5), 
    range(3))
# draw a line showing intersection with reports
map(lambda x: chart[2].axvline(x=rep_dates[x]-pd.Timedelta(12,'D'),ymin=-1,ymax=0.9,linestyle='--',color='gray',alpha=0.5), 
    range(3))
map(lambda x: chart[2].fill_between(x=pd.date_range(rep_dates[x]-pd.Timedelta(12,'D'),rep_dates[x]),
                                    y1=-2,y2=1.5,color='gray',alpha=0.2),
    range(3))
# show date of reports
map(lambda x: chart[2].text(rep_dates[x],-1,r'$\blacktriangle$',fontsize=50,color='orange',horizontalalignment='center'),
    range(3))
map(lambda x: chart[2].text(rep_dates[x],-0.95,r'$!$',fontsize=20,color='black',horizontalalignment='center'),
    range(3))
# show full date range
_ = chart[2].set_xlim('2050-05-01','2050-08-31')
# adjust y-axis span
_ = chart[2].set_ylim(-2,2)
# set the dates between the two records
_ = chart[2].spines['bottom'].set_position('center')
# only label values in three places
_ = chart[2].set_yticks([-1,0,1])
# name indicated data
_ = chart[2].set_yticklabels(['reports','dates','warm events'],fontsize=14)
# label number of disasters detected
_ = chart[2].text('2050-06-07',1.2,'Delay threshold = 12 days', style='italic',
                  bbox={'facecolor':c[4], 'alpha':0.4, 'pad':20})
_ = chart[2].text('2050-06-07',0.75,'Disasters detected = 3', style='italic',
                  bbox={'facecolor':c[4], 'alpha':0.4, 'pad':20})
# remove left axis line, add top line
sns.despine(bottom=False,left=True,top=False,ax=chart[2])
# set color of dividing line
chart[2].spines['top'].set_color(c[0])

# save figure
plt.savefig('../figures/reporting_example.png')
```
